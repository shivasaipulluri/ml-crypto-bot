```python
import time
import csv
import json
from coinbase.rest import RESTClient
from json import dumps

# Initialize the Coinbase client
api_key = "organizations/34ea76db-0149-4b56-a708-2610a3c60ab2/apiKeys/5533fcd1-0647-4784-a9b5-902bf53691c4"
api_secret = "-----BEGIN EC PRIVATE KEY-----\nMHcCAQEEIHshL3W7qRla2s+0t0ttn5u8ys70j3eJTwpk50uYCIQ6oAoGCCqGSM49\nAwEHoUQDQgAEj2YiJKTFGzUeSgcDVpgaAuBV4N9//6ZJ+Li0GdPaBo8Hd0yGQTLc\nWPGHUaot5a6HklujIN/zt21EgpBERDgl2Q==\n-----END EC PRIVATE KEY-----\n"

client = RESTClient(api_key=api_key, api_secret=api_secret)

# Define trading parameters
trade_amount = 5  # Amount in USD per trade
target_profit = 0.03  # 3% target profit
stop_loss_tolerance = -0.15  # 15% stop-loss
maker_fee = 0.004  # 0.4% maker fee
taker_fee = 0.006  # 0.6% taker fee

# File paths for logs
price_log_file = "btc_prices.csv"
trade_log_file = "trade_logs.csv"
stats_log_file = "stats_logs.csv"

# Initialize historical prices and logs
historical_prices = []
with open(price_log_file, "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["timestamp", "price", "moving_average", "volatility"])  # BTC price logs

with open(trade_log_file, "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["timestamp", "action", "price", "quantity", "profit", "reason"])  # Trade logs

with open(stats_log_file, "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["timestamp", "average_price", "min_price", "max_price", "volatility"])  # Stats logs

def log_price(btc_price, moving_avg, volatility):
    """Log BTC prices, moving average, and volatility to a CSV file."""
    with open(price_log_file, "a", newline="") as file:
        writer = csv.writer(file)
        writer.writerow([time.time(), btc_price, moving_avg, volatility])

def log_trade(action, price, quantity, profit, reason):
    """Log trade actions (buy/sell/stop-loss) to a CSV file."""
    with open(trade_log_file, "a", newline="") as file:
        writer = csv.writer(file)
        writer.writerow([time.time(), action, price, quantity, profit, reason])

def log_stats(average_price, min_price, max_price, volatility):
    """Log statistical metrics for further analysis."""
    with open(stats_log_file, "a", newline="") as file:
        writer = csv.writer(file)
        writer.writerow([time.time(), average_price, min_price, max_price, volatility])

def fetch_btc_price():
    """Fetch the current BTC price with a timeout."""
    try:
        print("Fetching BTC price...")
        product = client.get_product("BTC-USD", timeout=10)  # Timeout after 10 seconds
        print(f"BTC price fetched successfully: {product['price']}")
        return float(product["price"])
    except Exception as e:
        print(f"Error fetching BTC price: {e}")
        raise

def calculate_stats(prices):
    """Calculate moving average, min, max, and volatility."""
    moving_avg = sum(prices) / len(prices)
    min_price = min(prices)
    max_price = max(prices)
    volatility = max_price - min_price
    return moving_avg, min_price, max_price, volatility

def place_limit_buy(price):
    """Place a limit-buy order."""
    try:
        print(f"Placing limit buy order at ${price}...")
        order = client.limit_order_gtc_buy(
            client_order_id="BUY_" + str(int(time.time())),
            product_id="BTC-USD",
            quote_size=str(trade_amount),
            limit_price=str(price)
        )
        if order['success']:
            print(f"Limit buy order placed successfully: {order['success_response']['order_id']}")
            return order['success_response']['order_id']
        else:
            print(f"Limit buy order failed: {order['error_response']}")
            return None
    except Exception as e:
        print(f"Error placing limit buy order: {e}")
        raise

def place_limit_sell(price, size):
    """Place a limit-sell order."""
    try:
        print(f"Placing limit sell order at ${price}...")
        order = client.limit_order_gtc_sell(
            client_order_id="SELL_" + str(int(time.time())),
            product_id="BTC-USD",
            base_size=size,
            limit_price=str(price)
        )
        if order['success']:
            print(f"Limit sell order placed successfully.")
            return True
        else:
            print(f"Limit sell order failed: {order['error_response']}")
            return False
    except Exception as e:
        print(f"Error placing limit sell order: {e}")
        raise

active_trade = None  # Track active trade

def scalping_strategy():
    global active_trade
    print("Starting Scalping Strategy...")
    
    while True:
        try:
            # Step 1: Fetch BTC Price
            btc_price = fetch_btc_price()
            
            # Step 2: Update Historical Prices
            historical_prices.append(btc_price)
            if len(historical_prices) > 100:
                historical_prices.pop(0)
            
            # Step 3: Calculate Statistics
            moving_avg, min_price, max_price, volatility = calculate_stats(historical_prices)
            print(f"Moving Average: ${moving_avg:.2f}, Volatility: ${volatility:.2f}")
            log_price(btc_price, moving_avg, volatility)
            log_stats(moving_avg, min_price, max_price, volatility)
            
            # Step 4: Trading Logic
            if not active_trade:
                # Buy Condition
                if btc_price < moving_avg * (1 - maker_fee):
                    print("Buy condition met. Placing order...")
                    order_id = place_limit_buy(btc_price)
                    if order_id:
                        btc_size = trade_amount / btc_price
                        active_trade = {"buy_price": btc_price, "btc_size": btc_size, "order_id": order_id}
                        log_trade("BUY", btc_price, btc_size, 0, "Signal Detected")
            else:
                # Sell or Stop-Loss
                buy_price = active_trade["buy_price"]
                target_sell_price = buy_price * (1 + target_profit)
                stop_loss_price = buy_price * (1 + stop_loss_tolerance)
                
                if btc_price >= target_sell_price:
                    print("Sell condition met. Placing order...")
                    profit = (target_sell_price - buy_price) * active_trade["btc_size"]
                    if place_limit_sell(target_sell_price, active_trade["btc_size"]):
                        log_trade("SELL", btc_price, active_trade["btc_size"], profit, "Profit Target Reached")
                        active_trade = None
                elif btc_price <= stop_loss_price:
                    print("Stop-loss condition met. Placing order...")
                    loss = (buy_price - stop_loss_price) * active_trade["btc_size"]
                    if place_limit_sell(stop_loss_price, active_trade["btc_size"]):
                        log_trade("STOP-LOSS", btc_price, active_trade["btc_size"], -loss, "Stop Loss Triggered")
                        active_trade = None
            
            # Step 5: Wait Before Next Iteration
            time.sleep(10)
        
        except Exception as e:
            print(f"Error in main loop: {e}")
            time.sleep(30)  # Pause on error
```

```python
import time
import csv
import json
from coinbase.rest import RESTClient
from json import dumps

# Initialize the Coinbase client
api_key = "organizations/34ea76db-0149-4b56-a708-2610a3c60ab2/apiKeys/5533fcd1-0647-4784-a9b5-902bf53691c4"
api_secret = "-----BEGIN EC PRIVATE KEY-----\nMHcCAQEEIHshL3W7qRla2s+0t0ttn5u8ys70j3eJTwpk50uYCIQ6oAoGCCqGSM49\nAwEHoUQDQgAEj2YiJKTFGzUeSgcDVpgaAuBV4N9//6ZJ+Li0GdPaBo8Hd0yGQTLc\nWPGHUaot5a6HklujIN/zt21EgpBERDgl2Q==\n-----END EC PRIVATE KEY-----\n"

client = RESTClient(api_key=api_key, api_secret=api_secret)

# Define trading parameters
trade_amount = 5  # Amount in USD per trade
target_profit = 0.02  # 2% target profit
stop_loss_tolerance = -0.05  # 5% stop-loss
maker_fee = 0.004  # 0.4% maker fee
taker_fee = 0.006  # 0.6% taker fee

# File paths for logs
price_log_file = "btc_prices.csv"
trade_log_file = "trade_logs.csv"
stats_log_file = "stats_logs.csv"

# Initialize historical prices and logs
historical_prices = []
with open(price_log_file, "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["timestamp", "price", "moving_average", "volatility", "rsi", "atr"])  # BTC price logs

with open(trade_log_file, "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["timestamp", "action", "price", "quantity", "profit", "reason"])  # Trade logs

with open(stats_log_file, "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["timestamp", "average_price", "min_price", "max_price", "volatility", "rsi", "atr"])  # Stats logs

def log_price(btc_price, moving_avg, volatility, rsi, atr):
    """Log BTC prices, moving average, volatility, RSI, and ATR to a CSV file."""
    with open(price_log_file, "a", newline="") as file:
        writer = csv.writer(file)
        writer.writerow([time.time(), btc_price, moving_avg, volatility, rsi, atr])

def log_trade(action, price, quantity, profit, reason):
    """Log trade actions (buy/sell/stop-loss) to a CSV file."""
    with open(trade_log_file, "a", newline="") as file:
        writer = csv.writer(file)
        writer.writerow([time.time(), action, price, quantity, profit, reason])

def log_stats(average_price, min_price, max_price, volatility, rsi, atr):
    """Log statistical metrics for further analysis."""
    with open(stats_log_file, "a", newline="") as file:
        writer = csv.writer(file)
        writer.writerow([time.time(), average_price, min_price, max_price, volatility, rsi, atr])

def fetch_btc_price():
    """Fetch the current BTC price with a timeout."""
    try:
        product = client.get_product("BTC-USD", timeout=10)
        return float(product["price"])
    except Exception as e:
        print(f"Error fetching BTC price: {e}")
        raise

def calculate_stats(prices):
    """Calculate moving average, min, max, volatility, RSI, and ATR."""
    moving_avg = sum(prices) / len(prices)
    min_price = min(prices)
    max_price = max(prices)
    volatility = max_price - min_price

    # Calculate RSI
    gains = [prices[i + 1] - prices[i] for i in range(len(prices) - 1) if prices[i + 1] > prices[i]]
    losses = [-1 * (prices[i + 1] - prices[i]) for i in range(len(prices) - 1) if prices[i + 1] < prices[i]]
    avg_gain = sum(gains) / 14 if len(gains) >= 14 else sum(gains) / max(len(gains), 1)
    avg_loss = sum(losses) / 14 if len(losses) >= 14 else sum(losses) / max(len(losses), 1)
    rs = avg_gain / avg_loss if avg_loss != 0 else 0
    rsi = 100 - (100 / (1 + rs))

    # Calculate ATR
    tr_values = [max(prices[i] - prices[i - 1], abs(prices[i] - min_price), abs(prices[i] - max_price)) for i in range(1, len(prices))]
    atr = sum(tr_values) / len(tr_values) if tr_values else 0

    return moving_avg, min_price, max_price, volatility, rsi, atr

def place_limit_buy(price):
    """Place a limit-buy order."""
    try:
        order = client.limit_order_gtc_buy(
            client_order_id="BUY_" + str(int(time.time())),
            product_id="BTC-USD",
            quote_size=str(trade_amount),
            limit_price=str(price)
        )
        return order['success_response']['order_id'] if order['success'] else None
    except Exception as e:
        print(f"Error placing limit buy order: {e}")
        raise

def place_limit_sell(price, size):
    """Place a limit-sell order."""
    try:
        order = client.limit_order_gtc_sell(
            client_order_id="SELL_" + str(int(time.time())),
            product_id="BTC-USD",
            base_size=size,
            limit_price=str(price)
        )
        return order['success']
    except Exception as e:
        print(f"Error placing limit sell order: {e}")
        raise

active_trade = None  # Track active trade

def scalping_strategy():
    global active_trade
    print("Starting Scalping Strategy...")
    
    while True:
        try:
            # Step 1: Fetch BTC Price
            btc_price = fetch_btc_price()

            # Step 2: Update Historical Prices
            historical_prices.append(btc_price)
            if len(historical_prices) > 100:
                historical_prices.pop(0)

            # Step 3: Calculate Statistics
            moving_avg, min_price, max_price, volatility, rsi, atr = calculate_stats(historical_prices)
            print(f"BTC Price: {btc_price}, Moving Avg: {moving_avg}, RSI: {rsi}, ATR: {atr}")

            # Step 4: Log Data
            log_price(btc_price, moving_avg, volatility, rsi, atr)
            log_stats(moving_avg, min_price, max_price, volatility, rsi, atr)

            # Step 5: Trading Logic
            if not active_trade:
                # Buy Condition
                if btc_price < moving_avg * (1 - maker_fee) and rsi < 30:
                    print("Buy condition met. Placing buy order...")
                    order_id = place_limit_buy(btc_price)
                    if order_id:
                        btc_size = trade_amount / btc_price
                        active_trade = {"buy_price": btc_price, "btc_size": btc_size, "order_id": order_id}
                        log_trade("BUY", btc_price, btc_size, 0, "RSI & Moving Average")
            else:
                # Sell or Stop-Loss
                buy_price = active_trade["buy_price"]
                target_sell_price = buy_price * (1 + target_profit)
                stop_loss_price = buy_price * (1 + stop_loss_tolerance)

                if btc_price >= target_sell_price:
                    print("Sell condition met. Placing sell order...")
                    profit = (target_sell_price - buy_price) * active_trade["btc_size"]
                    if place_limit_sell(target_sell_price, active_trade["btc_size"]):
                        log_trade("SELL", btc_price, active_trade["btc_size"], profit, "Profit Target")
                        active_trade = None
                elif btc_price <= stop_loss_price:
                    print("Stop-loss condition met. Placing sell order...")
                    loss = (buy_price - stop_loss_price) * active_trade["btc_size"]
                    if place_limit_sell(stop_loss_price, active_trade["btc_size"]):
                        log_trade("STOP-LOSS", btc_price, active_trade["btc_size"], -loss, "Stop Loss")
                        active_trade = None

            # Step 6: Wait Before Next Iteration
            time.sleep(10)
        except Exception as e:
            print(f"Error in main loop: {e}")
            time.sleep(30)
```
